# Normal

```python
mu,sigma=0.9,0.8
s=np.random.normal(mu,sigma,700)
count,bins,ignored=plt.hist(s,200)

#distribution curve:
plt.plot(bins,1/sigma*np.sqrt(2*np.pi)*np.exp(-(bins-mu)**2/(2*sigma**2)))
plt.show()
```
